# １章 ウェブの注文数を分析する１０本ノック

ここでは、ある企業のECサイトでの商品の注文数の推移を分析していきます。  
データの属性を理解し、分析をするためにデータを加工した後、  
データの可視化を行うことで問題を発見していくプロセスを学びます。

### ノック１：データを読み込んでみよう

```python
import pandas as pd
customer_master = pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('item_master.csv')
item_master.head()
```

```python
transaction_1 = pd.read_csv('transaction_1.csv')
transaction_1.head()
```

```python
transaction_detail_1 = pd.read_csv('transaction_detail_1.csv')
transaction_detail_1.head()
```

### ノック２：データを結合(ユニオン)してみよう

```python
transaction_2 = pd.read_csv('transaction_2.csv')
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
transaction_detail_2 = pd.read_csv('transaction_detail_2.csv')
transaction_detail=pd.concat([transaction_detail_1,transaction_detail_2], ignore_index=True)
transaction_detail.head()
```

### ノック３：売上データ同士を結合(ジョイン)してみよう

```python
join_date = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]], on="transaction_id", how="left")
join_date.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_date))
```

### ノック４：マスタデータを結合(ジョイン)してみよう

```python
join_date = pd.merge(join_date, customer_master, on="customer_id", how="left")
join_date = pd.merge(join_date, item_master, on="item_id", how="left")
join_date.head()
```

### ノック5：必要なデータ列を作ろう

```python
join_date["price"] = join_date["quantity"] * join_date["item_price"]
join_date[["quantity", "item_price","price"]].head()
```

### ノック6：データ検算をしよう

```python
print(join_date["price"].sum())
print(transaction["price"].sum())
```

```python
join_date["price"].sum() == transaction["price"].sum()
```

### ノック7：各種統計量を把握しよう

```python
join_date.isnull().sum()
```

```python
join_date.describe()
```

```python
print(join_date["payment_date"].min())
print(join_date["payment_date"].max())
```

### ノック8：月別でデータを集計してみよう

```python
join_date.dtypes
```

```python
join_date["payment_date"] = pd.to_datetime(join_date["payment_date"])
join_date["payment_month"] = join_date["payment_date"].dt.strftime("%Y%m")
join_date[["payment_date", "payment_month"]].head()
```

```python
join_date.groupby("payment_month").sum()["price"]
```

### ノック9：月別、商品別でデータを集計してみよう

```python
join_date.groupby(["payment_month", "item_name"]).sum()[["price", "quantity"]]
```

```python
pd.pivot_table(join_date, index='item_name', columns='payment_month', values=['price','quantity'], aggfunc='sum')
```

### ノック10：商品別の売上推移を可視化してみよう

```python
graph_date = pd.pivot_table(join_date, index='payment_month', columns='item_name', values='price', aggfunc='sum')
graph_date.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(list(graph_date.index),graph_date["PC-A"], label='PC-A')
plt.plot(list(graph_date.index),graph_date["PC-B"], label='PC-B')
plt.plot(list(graph_date.index),graph_date["PC-C"], label='PC-C')
plt.plot(list(graph_date.index),graph_date["PC-D"], label='PC-D')
plt.plot(list(graph_date.index),graph_date["PC-E"], label='PC-E')
plt.legend()
```
