# ２章　小売店のデータでデータ加工を行う１０本ノック

本章では、ある小売店の売上履歴と顧客台帳データを用いて、データ分析の素地となる「データの加工」を習得することが目的です。
実際の現場データは手入力のExcel等、決して綺麗なデータではない事が多いため、
データの揺れや整合性の担保など、汚いデータを取り扱うデータ加工を主体に進めて行きます。

### ノック１１：データを読み込んでみよう

```python
import pandas as pd
uriage_date = pd.read_csv("uriage.csv")
uriage_date.head()
```

```python
kokyaku_date = pd.read_excel("kokyaku_daicho.xlsx")
kokyaku_date.head()
```

### ノック１２：データの揺れを見てみよう

```python
uriage_date["item_name"].head()
```

```python
uriage_date["item_price"].head()
```

### ノック１３：データに揺れがあるまま集計しよう

```python
uriage_date["purchase_date"] = pd.to_datetime(uriage_date["purchase_date"])
uriage_date["purchase_month"] = uriage_date["purchase_date"].dt.strftime("%Y%m")
res = uriage_date.pivot_table(index="purchase_month", columns="item_name", aggfunc="size", fill_value=0)
res
```

```python
res = uriage_date.pivot_table(index="purchase_month", columns="item_name", values="item_price", aggfunc="sum", fill_value=0)
res
```

### ノック１４：商品名の揺れを補正しよう

```python
print(len(pd.unique(uriage_date["item_name"])))
```

```python
uriage_date["item_name"] = uriage_date["item_name"].str.upper()
uriage_date["item_name"] = uriage_date["item_name"].str.replace("　", "")
uriage_date["item_name"] = uriage_date["item_name"].str.replace(" ", "")
uriage_date.sort_values(by=["item_name"], ascending=True)
```

```python
print(pd.unique(uriage_date["item_name"]))
print(len(pd.unique(uriage_date["item_name"])))
```

### ノック１５：金額欠損値の補完をしよう

```python
uriage_date.isnull().any(axis=0)
```

```python
flg_is_null = uriage_date["item_price"].isnull()
for trg in list(uriage_date.loc[flg_is_null, "item_name"].unique()):
    price = uriage_date.loc[(~flg_is_null) & (uriage_date["item_name"] == trg), "item_price"].max()
    uriage_date.loc[(flg_is_null) & (uriage_date["item_name"]==trg), "item_price"] = price
uriage_date.head()
```

```python
uriage_date.isnull().any(axis=0)
```

```python
for trg in list(uriage_date["item_name"].sort_values().unique()):
  print(trg + "の最大額:" + str(uriage_date.loc[uriage_date["item_name"]==trg]["item_price"].max()) + "の最小額:" + str(uriage_date.loc[uriage_date["item_name"]==trg]["item_price"].min(skipna=False)))
```

### ノック１６：顧客名の揺れを補正しよう

```python
kokyaku_date["顧客名"].head()
```

```python
uriage_date["customer_name"].head()
```

```python
kokyaku_date["顧客名"] = kokyaku_date["顧客名"].str.replace("　", "")
kokyaku_date["顧客名"] = kokyaku_date["顧客名"].str.replace(" ", "")
kokyaku_date["顧客名"].head()
```

### ノック１７：日付の揺れを補正しよう

```python
flg_is_serial = kokyaku_date["登録日"].astype("str").str.isdigit()
flg_is_serial.sum()
```

```python
fromSerial = pd.to_timedelta(kokyaku_date.loc[flg_is_serial,"登録日"].astype("float") - 2, unit="D") + pd.to_datetime('1990/1/1')
fromSerial
```

```python
fromString = pd.to_datetime(kokyaku_date.loc[~flg_is_serial, "登録日"])
fromString
```

```python
kokyaku_date["登録日"] = pd.concat([fromSerial, fromString])
kokyaku_date
```

```python
kokyaku_date["登録年月"] = kokyaku_date["登録日"].dt.strftime("%Y%m")
rslt = kokyaku_date.groupby("登録年月").count()["顧客名"]
print(rslt)
print(len(kokyaku_date))
```

```python
flg_is_serial = kokyaku_date["登録日"].astype("str").str.isdigit()
flg_is_serial.sum()
```

### ノック１８：顧客名をキーに２つのデータを結合(ジョイン)しよう

```python
join_data = pd.merge(uriage_date, kokyaku_date, left_on="customer_name", right_on="顧客名", how="left")
join_data = join_data.drop("customer_name", axis=1)
join_data
```

### ノック１９：クレンジングしたデータをダンプしよう

```python
dump_data = join_data[["purchase_date", "purchase_month", "item_name", "item_price", "顧客名", "かな", "地域", "メールアドレス", "登録日"]]
dump_data
```

```python
dump_data.to_csv("dump_data.csv", index=False)
```
